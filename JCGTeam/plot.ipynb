# Fine Tuning

## 0%-30% Freeze

```python
freeze_amount = [0, 5, 10, 15, 20, 25, 30]

zeroToThirty = pickle.load(open("Histories/FreezeRange/history_list0to30.p", "rb"))
zeroToThirtyMeta = pickle.load(open("Histories/FreezeRange/metadatahistory_list0to30.p", "rb"))

training_history = []
pretraining_history = []

for freeze in range(len(zeroToThirty)):
    training_history.append(zeroToThirty[freeze][0])
    pretraining_history.append(zeroToThirty[freeze][1])

full_history_val = [y['val_accuracy']+x['val_accuracy'] for x,y in zeroToThirty]
full_history_acc = [y['accuracy']+x['accuracy'] for x,y in zeroToThirty]

plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['red', 'green', 'blue', 'yellow', 'olive', 'cyan', 'magenta'])
plt.figure(figsize=(10,5))
for i in range(len(training_history)):
    print("Maximum validation accuracy for freeze amount of", freeze_amount[i], "is", max(full_history_val[i]))
    plt.plot(full_history_acc[i],alpha=.3, linestyle='--', linewidth=3)
    plt.plot(full_history_val[i],  label = str(freeze_amount[i])+"%", alpha=.8, linewidth=3)
    plt.legend(loc='lower center',  ncol=4, prop={'size': 18})
    plt.title('Accuracy for Different Freeze Amounts (0%-30%)', fontsize=20)
    plt.grid(alpha=.3)
plt.show()
```
