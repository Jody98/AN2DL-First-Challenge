```python
import pickle
import matplotlib.pyplot as plt
```

```python
# load history from pickle file
historyInceptionV3 = pickle.load(open("history_InceptionV3.p", "rb"))
historyVGG16 = pickle.load(open("history_VGG16.p", "rb"))
historyXception = pickle.load(open("history_Xception.p", "rb"))
historyDenseNet201 = pickle.load(open("history_DenseNet201.p", "rb"))
historyDenseNet121 = pickle.load(open("history_DenseNet121.p", "rb"))
historyDenseNet169 = pickle.load(open("history_DenseNet169.p", "rb"))
historyEfficientNetB3 = pickle.load(open("history_efnetB3.p", "rb"))
historyEfficientNetB4 = pickle.load(open("history_efnetB4.p", "rb"))
historyEfficientNetB5 = pickle.load(open("history_efnetB5.p", "rb"))

histories = [historyInceptionV3, historyVGG16, historyXception, historyDenseNet201, historyDenseNet121, historyDenseNet169, historyEfficientNetB3, historyEfficientNetB4, historyEfficientNetB5]
```

```python
# Plot the training history
plt.figure(figsize=(10,5))
""""""""""""""""
Categorical Crossentropy
"""""""""""""""""
# InceptionV3
plt.plot(historyInceptionV3['loss'], alpha=.3, color='r', linestyle='--', linewidth=3)
plt.plot(historyInceptionV3['val_loss'], label='InceptionV3', alpha=.8, color='r', linewidth=3)
#VGG16
plt.plot(historyVGG16['loss'], alpha=.3, color='g', linestyle='--', linewidth=3)
plt.plot(historyVGG16['val_loss'], label='VGG16', alpha=.8, color='g', linewidth=3)
#Xception
plt.plot(historyXception['loss'], alpha=.3, color='darkorange', linestyle='--', linewidth=3)
plt.plot(historyXception['val_loss'], label='Xception', alpha=.8, color='darkorange', linewidth=3)
# DenseNet201
plt.plot(historyDenseNet201['loss'], alpha=.3, color='deeppink', linestyle='--', linewidth=3)
plt.plot(historyDenseNet201['val_loss'], label='DenseNet201', alpha=.8, color='deeppink', linewidth=3)
#DenseNet121
plt.plot(historyDenseNet121['loss'], alpha=.3, color='orchid', linestyle='--', linewidth=3)
plt.plot(historyDenseNet121['val_loss'], label='DenseNet121', alpha=.8, color='orchid', linewidth=3)
#DenseNet169
plt.plot(historyDenseNet169['loss'], alpha=.3, color='purple', linestyle='--', linewidth=3)
plt.plot(historyDenseNet169['val_loss'], label='DenseNet169', alpha=.8, color='purple', linewidth=3)
# EfficientNetB3
plt.plot(historyEfficientNetB3['loss'], alpha=.3, color='steelblue', linestyle='--', linewidth=3)
plt.plot(historyEfficientNetB3['val_loss'], label='EfficientNetB3', alpha=.8, color='steelblue', linewidth=3)
# EfficientNetB4
plt.plot(historyEfficientNetB4['loss'], alpha=.3, color='orchid', linestyle='--', linewidth=3)
plt.plot(historyEfficientNetB4['val_loss'], label='EfficientNetB4', alpha=.8, color='orchid', linewidth=3)
# EfficientNetB5
plt.plot(historyEfficientNetB5['loss'], alpha=.3, color='dodgerblue', linestyle='--', linewidth=3)
plt.plot(historyEfficientNetB5['val_loss'], label='EfficientNetB5', alpha=.8, color='dodgerblue', linewidth=3)

plt.legend(loc='upper right', prop={'size': 18})
plt.title('Categorical Crossentropy', fontsize=20)
plt.grid(alpha=.3)

""""""""""""""""
Accuracy
"""""""""""""""""
plt.figure(figsize=(10,5))

plt.plot(historyInceptionV3['accuracy'], label='Training Accuracy', alpha=.3, color='r', linestyle='--', linewidth=3)
plt.plot(historyInceptionV3['val_accuracy'], label='Validation Accuracy', alpha=.8, color='r', linewidth=3)

# InceptionV3
plt.plot(historyInceptionV3['accuracy'], alpha=.3, color='r', linestyle='--', linewidth=3)
plt.plot(historyInceptionV3['val_accuracy'], label='InceptionV3', alpha=.8, color='r', linewidth=3)
#VGG16
plt.plot(historyVGG16['accuracy'], alpha=.3, color='g', linestyle='--', linewidth=3)
plt.plot(historyVGG16['val_accuracy'], label='VGG16', alpha=.8, color='g', linewidth=3)
#Xception
plt.plot(historyXception['accuracy'], alpha=.3, color='darkorange', linestyle='--', linewidth=3)
plt.plot(historyXception['val_accuracy'], label='Xception', alpha=.8, color='darkorange', linewidth=3)
# DenseNet201
plt.plot(historyDenseNet201['accuracy'], alpha=.3, color='deeppink', linestyle='--', linewidth=3)
plt.plot(historyDenseNet201['val_accuracy'], label='DenseNet201', alpha=.8, color='deeppink', linewidth=3)
#DenseNet121
plt.plot(historyDenseNet121['accuracy'], alpha=.3, color='orchid', linestyle='--', linewidth=3)
plt.plot(historyDenseNet121['val_accuracy'], label='DenseNet121', alpha=.8, color='orchid', linewidth=3)
#DenseNet169
plt.plot(historyDenseNet169['accuracy'], alpha=.3, color='purple', linestyle='--', linewidth=3)
plt.plot(historyDenseNet169['val_accuracy'], label='DenseNet169', alpha=.8, color='purple', linewidth=3)
# EfficientNetB3
plt.plot(historyEfficientNetB3['accuracy'], alpha=.3, color='steelblue', linestyle='--', linewidth=3)
plt.plot(historyEfficientNetB3['val_accuracy'], label='EfficientNetB3', alpha=.8, color='steelblue', linewidth=3)
# EfficientNetB4
plt.plot(historyEfficientNetB4['accuracy'], alpha=.3, color='orchid', linestyle='--', linewidth=3)
plt.plot(historyEfficientNetB4['val_accuracy'], label='EfficientNetB4', alpha=.8, color='orchid', linewidth=3)
# EfficientNetB5
plt.plot(historyEfficientNetB5['accuracy'], alpha=.3, color='dodgerblue', linestyle='--', linewidth=3)
plt.plot(historyEfficientNetB5['val_accuracy'], label='EfficientNetB5', alpha=.8, color='dodgerblue', linewidth=3)

plt.legend(loc='lower right', prop={'size': 18})
plt.title('Accuracy', fontsize=20)
plt.grid(alpha=.3)

plt.show()
```
